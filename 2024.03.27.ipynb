```python
import datetime
import time

def calc(datas):
    mean = 0
    for data in datas:
        if data >= 900000:
            data = data - 1000000
        mean = mean + data
    
    return mean / len(datas)


case_1 = []

for i in range(50):
    # microsecond 0 / compare datetime 
    base = datetime.datetime.now().replace(microsecond=0) + datetime.timedelta(seconds=2)
    while True:
        now = datetime.datetime.now()
        if now >= base:
            break
        else:
            time.sleep(0.001)
    case_1.append(now.microsecond)
    print('case_1 : ' + str(i))
print('case_1 done')
print('1 result : {0}'.format(calc(case_1)))
```
