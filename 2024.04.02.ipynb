```python
def center(s):
    글자길이 = len(s)
    중간번호 = 글자길이 // 2
    if 글자길이 % 2 == 0:
        print(s[중간번호-1:중간번호+1])
    else:
        print(s[중간번호])
center('125656')
```

```python
def 수박 (n):
    result = '' #수박 글가를 담을 변수
    for i in range(n):
        if i % 2 == 0:
            result += '수'
        else:
            result += '박' 
    return result
수박(6)
```
