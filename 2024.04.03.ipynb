```python
import requests
res = requests.get('https://httpstatusdogs.com/')
t = res.text
```

```python
import re
규칙 = re.compile('img/[0-9]{3}.jpg')
결과 = 규칙.findall(t)
s = set(결과)
for a in s:
    res = requests.get(f'https://httpstatusdogs.com/{a}')
    f = open(a.split('/')[1],'wb')
    f.write(res.content)
    f.close()
```
