```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# 브라우저 드라이버 초기화
driver = webdriver.Chrome()  # Chrome 브라우저를 사용합니다.

# 원하는 URL로 이동
url = 'https://www.etnews.com/news/section.html?id1=02&page=3'
driver.get(url)

# 원하는 페이지 번호
desired_page = 7

# 현재 페이지 번호를 나타내는 변수 (초기값을 0으로 설정)
current_page = 3

# 반복문을 통해 원하는 페이지까지 이동합니다.
while current_page != desired_page:
    # 현재 페이지 번호 요소를 찾습니다.
    current_page_element = driver.find_element(By.CSS_SELECTOR, 'a')
    current_page = int(current_page_element.text)

    # 원하는 페이지에 도달했는지 확인합니다.
    if current_page == desired_page:
        break

    # 다음 페이지로 이동하는 요소를 찾아 클릭합니다.
    next_page_element = driver.find_element(By.CSS_SELECTOR, 'a')
    next_page_element.click()

    # 페이지 로드 대기
    time.sleep(1)

# 원하는 페이지에 도달한 후 작업을 수행합니다.
# main_news_img 클래스를 가진 요소를 찾습니다.
elem = driver.find_element(By.CLASS_NAME, "main_news_img")

# main_news_img 요소 내에서 태그 이름이 "a"인 요소를 찾습니다.
news_elem = elem.find_elements(By.TAG_NAME, " article:nth-child(1) ")

# 첫 번째 "a" 요소를 클릭합니다.
news_elem[0].click()

# 이전 페이지로 돌아갑니다.
driver.execute_script("window.history.go(-1)")

# 브라우저 종료
driver.quit()
```

```python
def insert_nara(dt):
    sql = "insert into nara(public_no,public_name,page_url)"
    sql += f" values('{dt['public_no']}', '{dt['public_name']}', '{dt['page_url']}')"
    conn.execute(text(sql))
    conn.execute(text('commit'))
    return pd.read_sql(f"select * from nara where public_no='{dt['public_no']}'",conn)  
```

```python
def update_nara(public_no,*args):
    sql = "update nara set "
    num = len(args)
    i = 1
    for col,val in args:
        if i < num:
            sql += f",{col}= '{val}', "
        else:
            sql += f",{col}= '{val}' "
        i += 1
    sql += f"where public_no = '{public_no}'"
    conn.execute(text(sql))
    conn.execute(text('commit'))
    return pd.read_sql("select * from nara,conn")
```
