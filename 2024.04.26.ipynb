```python
score = np.random.randint(60,101,10_000_000)
```

```python
import datetime
##90이상 A,80이상 B, 70이상 c ,나머지 D
curr = datetime.datetime.now()
score_class = []
for s in score:
    if s >= 90:
        score_class.append('A')
    elif s >= 80:
        score_class.append('B')
    elif s >= 70:
        score_class.append('C')
    else:
        score_class.append('D')

elpase_time = datetime.datetime.now()- curr
elpase_time
```

```python
score_class = np.where(score >=90,'A',
                       np.where(score >=80,'b',
                                np.where(score >70,'C','D')))
elpase_time = datetime.datetime.now()- curr
elpase_time 
# np.where() 함수를 사용한 방법이 훨씬 빠르다
```
